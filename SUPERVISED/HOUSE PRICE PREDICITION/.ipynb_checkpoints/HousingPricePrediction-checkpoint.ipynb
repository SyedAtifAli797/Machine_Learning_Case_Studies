                                                COVERSION OF CATEGORICAL COLUMN INTO NUMERICAL

--> BY TAKING OUT UNIQUE VALUES TO PERFORM ONE-HOT ENCODING:-

```python
#gives total unique values in the column:-
df['ocean_proximity'].unique()
```

```python
#gives number of unique values:-
df['ocean_proximity'].nunique()
```

```python
#counts all the occurrence of all unique values:-
df['ocean_proximity'].value_counts()
```

```python
#series.values returns 1D array:-
df['ocean_proximity'].values
```

```python
# len function is used to check the length of an array
len(df['ocean_proximity'].values)
```

```python
# Since one hot ecoding needs a 2D, therefore,
# we will convert the series into a 2D array

categorical_data_array = df['ocean_proximity'].values
```

```python
#reshape method is used to reshape the data into 2D array
array_2d = categorical_data_array.reshape(-1,1)
```

```python
array_2d
```

--> Importing one-hot ecoder from the scikit-learn's preprocessing class 

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded_data = encoder.fit_transform(array_2d)
```

```python
encoded_data
```

```python
encoded_data.toarray()
```

```python
encoder.categories_[0]
```

```python
df[encoder.categories_[0]] = encoded_data.toarray()
```

```python
df.head()
```

```python
df.drop('ocean_proximity', axis=1, inplace=True)
```

```python
df.head(2)
```

                                                        SPLITTING THE DATASET INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
```

```python
# df.iloc[0,5], exculde the last values
# df.loc[] - in this column index is the name of the column itself, does not exclude last values
```

```python
# We will split the dataset into 80% and 20%
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

--> VISUALIZING THE GRAPH OF THE ACTUAL TRAINING DATA VS PREDICTED DATA                                                                 

```python
import numpy as np
```

```python

plt.plot(np.array(y_train)[:200], label = 'actual')
plt.plot(y_pred[:200], label = 'predicted')
plt.legend()
plt.xlabel('Number of rows')
plt.ylabel('Total House Value')
plt.show()
```

```python
reg_rmse = mean_squared_error(y_train, y_pred, squared = False)
reg_rmse
```

                                                GIVING THE TRAINING DATA TO THE MODEL- RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest_reg = RandomForestRegressor(n_estimators = 100, max_features = 10, random_state = 42)
forest_reg.fit(X_train, y_train)
```

```python
y_pred_3 = forest_reg.predict(X_train)
y_pred_3
```

```python
plt.plot(np.array(y_train)[:200], label='actual')
plt.plot(y_pred_3[:200], label = 'predicted')
plt.legend()
plt.xlabel('Number of Rows')
plt.ylabel('Total House Value')
plt.show()
```

```python
sns.kdeplot(y_train, label='actual')
sns.kdeplot(y_pred_3, label='predicted')
plt.legend()
plt.xlabel('House Prices')
plt.ylabel('Density of Prices')
plt.show()
```

```python
forest_rmse = mean_squared_error(y_train, y_pred_3, squared = False)
forest_rmse
```

```python
from sklearn.model_selection import cross_val_score
```

```python
tree_reg_rmse_scores
```
