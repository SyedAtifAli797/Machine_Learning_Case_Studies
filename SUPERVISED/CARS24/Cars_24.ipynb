## Modeling

###  Decision Tree Regressor With Hyperparameter Tuning

```python
from Sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import KFold, cross_validate
```

```python

params={
  'max_depth':[5,10,15,20],
  'min_samples_split':[2,5,10,15],
  'min_samples_leaf':[1,2,4,5]
}
```

```python

randomcv = RandomizedSearchCV(estimator = DecisionTreeRegressor(random_state=7),
                            param_distributions = params,
                            scoring = 'r2',
                            cv = 3,
                            n_iter=15,
                            n_jobs=-1,
                            verbose=10
                            )
```

```python
randomcv.fit(X_train, y_train)
```

```python
print("Best params: ", randomcv.best_params_)
print("Best score: ", randomcv.best_score_)
```

```python
tree_model= DecisionTreeRegressor(random_state=7,
                                  min_samples_split=2,
                                  min_samples_leaf=5,
                                  max_depth=10
                          )

kfold = KFold(n_splits=10)
cv_acc_results = cross_validate(tree_model, X_train, y_train, cv=kfold, scoring='r2', return_train_score=True)
```

```python
print("K-Fold Accuracy Mean:")
print(f"Train: {cv_acc_results['train_score'].mean()*100}")
print(f"Validation: {cv_acc_results['test_score'].mean()*100}")
```

```python
tree_model.fit(X_train, y_train)
```

```python
print(f"Train Accuracy: {round(tree_model.score(X_train, y_train)*100,2)}")
print(f"Test Accuracy: {round(tree_model.score(X_test, y_test)*100,2)}")
```
