#### c. How do monthly or yearly sales trends look over time?

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.head()
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()
monthly_sales
```

```python
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
yearly_sales
```

```python
plt.figure(figsize=(14,5)).suptitle('Monthly and yearly sales')
plt.subplot(1,3,1)
sns.lineplot(x = monthly_sales.index , y = monthly_sales.values, color = 'blue')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()

plt.subplot(1,3,3)
sns.lineplot(x = yearly_sales.index , y = yearly_sales.values, color = 'red')
plt.xlabel('Years')
plt.xticks(rotation = 90)
plt.ylabel('Sales')
plt.grid()

plt.show()
```

#### d. What are the peak and low sales months for each year?

```python
monthly_sales_2019 = df.groupby(df['Order Date'].dt.year).get_group(2019).groupby(df['Order Date'].dt.month)['Sales'].sum()
monthly_sales_2019
```

```python
monthly_sales_2020 = df.groupby(df['Order Date'].dt.year).get_group(2020).groupby(df['Order Date'].dt.month)['Sales'].sum()
monthly_sales_2020
```

```python
monthly_sales_2021 = df.groupby(df['Order Date'].dt.year).get_group(2021).groupby(df['Order Date'].dt.month)['Sales'].sum()
monthly_sales_2021
```

```python
monthly_sales_2022 = df.groupby(df['Order Date'].dt.year).get_group(2022).groupby(df['Order Date'].dt.month)['Sales'].sum()
monthly_sales_2022
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x = monthly_sales_2019.index, y = monthly_sales_2019.values, label = '2019 Sales')
sns.lineplot(x = monthly_sales_2020.index, y = monthly_sales_2020.values, label = '2020 Sales')
sns.lineplot(x = monthly_sales_2021.index, y = monthly_sales_2021.values, label = '2021 Sales')
sns.lineplot(x = monthly_sales_2022.index, y = monthly_sales_2022.values, label = '2022 Sales')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x = monthly_sales_2019.index, y = monthly_sales_2019.values, label = '2019 Sales')
plt.text(2,min(monthly_sales_2019.values), 'Minimum sales of 2019', fontsize=8)
plt.text(9,max(monthly_sales_2019.values), 'Maximum sales of 2019', fontsize=8)

sns.lineplot(x = monthly_sales_2020.index, y = monthly_sales_2020.values, label = '2020 Sales')
plt.text(2,min(monthly_sales_2020.values), 'Minimum sales of 2020', fontsize=8)
plt.text(11,max(monthly_sales_2020.values), 'Maximum sales of 2020', fontsize=8)

sns.lineplot(x = monthly_sales_2021.index, y = monthly_sales_2021.values, label = '2021 Sales')
plt.text(1,min(monthly_sales_2021.values), 'Minimum sales of 2021', fontsize=8)
plt.text(12,max(monthly_sales_2021.values), 'Maximum sales of 2021', fontsize=8)

sns.lineplot(x = monthly_sales_2022.index, y = monthly_sales_2022.values, label = '2022 Sales')
plt.text(2,min(monthly_sales_2022.values), 'Minimum sales of 2022', fontsize=8)
plt.text(11,max(monthly_sales_2022.values), 'Maximum sales of 2022', fontsize=8)

plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 8))

sns.lineplot(x=monthly_sales_2019.index, y=monthly_sales_2019.values, label='2019 Sales', marker='o', color='blue')
plt.annotate('Minimum sales of 2019', 
             xy=(2, min(monthly_sales_2019.values)), 
             xytext=(4, min(monthly_sales_2019.values) - 2000), 
             arrowprops=dict(arrowstyle='->', color='blue'),
             fontsize=10, color='blue')
plt.annotate('Maximum sales of 2019', 
             xy=(9, max(monthly_sales_2019.values)), 
             xytext=(7, max(monthly_sales_2019.values) + 2500), 
             arrowprops=dict(arrowstyle='->', color='blue'),
             fontsize=10, color='blue')

sns.lineplot(x=monthly_sales_2020.index, y=monthly_sales_2020.values, label='2020 Sales', marker='o', color='green')
plt.annotate('Minimum sales of 2020', 
             xy=(2, min(monthly_sales_2020.values)), 
             xytext=(5, min(monthly_sales_2020.values) - 3000), 
             arrowprops=dict(arrowstyle='->', color='green'),
             fontsize=10, color='green')
plt.annotate('Maximum sales of 2020', 
             xy=(11, max(monthly_sales_2020.values)), 
             xytext=(9, max(monthly_sales_2020.values) + 12000), 
             arrowprops=dict(arrowstyle='->', color='green'),
             fontsize=10, color='green')

sns.lineplot(x=monthly_sales_2021.index, y=monthly_sales_2021.values, label='2021 Sales', marker='o', color='red')
plt.annotate('Minimum sales of 2021', 
             xy=(1, min(monthly_sales_2021.values)), 
             xytext=(3, min(monthly_sales_2021.values) - 4000), 
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=10, color='red')
plt.annotate('Maximum sales of 2021', 
             xy=(12, max(monthly_sales_2021.values)), 
             xytext=(10, max(monthly_sales_2021.values) + 4500), 
             arrowprops=dict(arrowstyle='->', color='red'),
             fontsize=10, color='red')

sns.lineplot(x=monthly_sales_2022.index, y=monthly_sales_2022.values, label='2022 Sales', marker='o', color='purple')
plt.annotate('Minimum sales of 2022', 
             xy=(2, min(monthly_sales_2022.values)), 
             xytext=(6, min(monthly_sales_2022.values) - 5000), 
             arrowprops=dict(arrowstyle='->', color='purple'),
             fontsize=10, color='purple')
plt.annotate('Maximum sales of 2022', 
             xy=(11, max(monthly_sales_2022.values)), 
             xytext=(9, max(monthly_sales_2022.values) ), 
             arrowprops=dict(arrowstyle='->', color='purple'),
             fontsize=10, color='purple')

plt.grid()
plt.legend(loc='upper left', fontsize=10)
plt.title("Monthly Sales Comparison (2019-2022)", fontsize=16)
plt.xlabel("Months", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.tight_layout()
plt.show()
```

# 3. Category-Based Analysis-

#### a. What are the top-selling categories and sub-categories?

```python
each_category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
each_category_sales[:1]
```

```python
sns.barplot(each_category_sales, color='red')
plt.show()
```

```python
each_sub_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
each_sub_sales[:1]
```

```python
plt.figure(figsize=(15,6))
sns.barplot(each_sub_sales, color='blue')
plt.xticks(rotation = 40)
plt.show()
```

#### b. How does the average order quantity vary across different categories?

```python
average_quantity_order = df.groupby('Category')['Quantity'].mean()
average_quantity_order
```

```python
sns.lineplot(average_quantity_order, color= 'green', marker='o')
plt.grid()
plt.show()
```
